```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC,SVR
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
data=load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['species']=data.target
df
```

```python
X=df.drop('species',axis=1)
y=df.species
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model=DecisionTreeClassifier(max_depth=5)
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(Xtest)
ypred
```

```python
print('Training Efficiency = ',model.score(Xtrain,ytrain))
print('Testing Efficiency = ',model.score(Xtest,ytest))
```

```python
print('Precision = ',precision_score(ytest,ypred,average='macro'))
print('Recall = ',recall_score(ytest,ypred,average=None))
print('Accuracy = ',accuracy_score(ytest,ypred))
```

```python

import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC,SVR
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,r2_score,mean_absolute_error

rdata=fetch_california_housing()
df=pd.DataFrame(rdata.data,columns=rdata.feature_names)
df['target']=rdata.target
df

X=df.drop('target',axis=1)
y=df.target

Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=42)

rmodel=DecisionTreeRegressor(max_depth=5)
rmodel.fit(Xtrain,ytrain)

yrpred=rmodel.predict(Xtest)
yrpred

print('Training Eff.= ',rmodel.score(Xtrain,ytrain))
print('Testing Eff.= ',rmodel.score(Xtest,ytest))

r2_score(ytest,yrpred)

mean_absolute_error(ytest,yrpred)
```
