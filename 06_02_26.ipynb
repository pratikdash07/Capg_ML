```python
# Original 3D array d3
print("Original d3 array:")
print(d3)

# 1. Access the second 2D array (index 1 along the first axis)
print("\nAccessing the second 2D array (d3[1]):")
print(d3[1])

# 2. Access the first row of all 2D arrays (across the second axis)
print("\nAccessing the first row across all 2D arrays (d3[:, 0, :]):")
print(d3[:, 0, :])

# 3. Access the last column of all 2D arrays (across the third axis)
print("\nAccessing the last column across all 2D arrays (d3[:, :, -1]):")
print(d3[:, :, -1])
# 4. Extract a sub-block: from the first 2D array, take the first two rows and first two columns
print("\nExtracting a sub-block (d3[0, :2, :2]):")
print(d3[0, :2, :2])

# 5. Extract elements using a step (e.g., every other column from all 2D arrays)
print("\nExtracting every other column from all 2D arrays (d3[:, :, ::2]):")
print(d3[:, :, ::2])

```
