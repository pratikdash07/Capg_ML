```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Data Cleaning

df=pd.read_csv("house_price.csv")
df.head()

df.tail()

df.columns

df.rename(columns={'House_ID':'HouseID','Owner_Name':'OwnerName'},inplace=True)

df.columns

df.dtypes

df.info()

df.isnull().sum()

df.describe()

df.describe(include='object')

(df['Year_Built']==2028).sum()

df['Notes'].isnull().sum()/(len(df['Notes']))*100
```

```python
df

df.drop(['OwnerName','Notes','Contact','Email'],axis=1)

df['Date_Sold']=pd.to_datetime(df['Date_Sold'],format = 'mixed', errors='coerce')
df['Date_Sold']

df['Date_Sold'].isnull().sum()
```

```python
# to extract day month and year

df['Day']=df['Date_Sold'].dt.day
df

df['Month']=df['Date_Sold'].dt.month

df['Year']=df['Date_Sold'].dt.year

df['Area_sqft'].unique()

df['Area_sqft']=df['Area_sqft'].str.extract('(\d+)')##Reg Expression

df['Area_sqft'].dtypes

df['Area_sqft'].isnull().sum()/len(df['Area_sqft'])*100

import seaborn as sns
plt.figure(figsize=(6,4))
sns.histplot(df['Area_sqft'].dropna(),kde=True,bins=100)

df['Area_sqft']=df['Area_sqft']
```

```python
df
```

```python
df['Bedrooms'] = df['Bedrooms'].replace('Three', '3')
df['Bedrooms'] = df['Bedrooms'].replace('Two', '2')
```

```python
df['City'] = df['City'].replace(['BBSR','Bhubaneshwar'],'Bhubaneswar')

df['Garage'] = df['Garage'].replace(['YES','Y','N'],['Yes','Yes','No'])
df['Garage'] = df['Garage'].replace({'Yes': '1', 'No':'0'})
df['Furnished'] = df['Furnished'].replace(['YES','Y','N'],['Yes','Yes','No'])
```

```python
df.isnull().sum()
```

```python

df['Floor_No'] = df['Floor_No'].replace(['Ground','G'],'0')
df['Floor_No'] = df['Floor_No'].astype(int)
df['Bedrooms'] = df['Bedrooms'].astype(int)
df['Garage'] = df['Garage'].astype(int)
df['Garage'] = df['Garage'].astype(bool)
df['Area_sqft'] = df['Area_sqft'].astype(float)

df
```

```python
df = df.drop(['OwnerName','Notes', 'Contact', 'Email'], axis=1)
df
```

```python
df.isnull().sum()
```

```python
#To convert price given as M(illions) to int
def convert_price(x):
  if isinstance(x,str) and 'M' in x:
    return float(x.replace('M',''))*1000000
  else:
    return float(x)
```

```python
df['Price'] = df['Price'].apply(convert_price)
df['Price']
```

```python
df.info()
```

```python
categoracal_cols = df.select_dtypes(include='object').columns
categoracal_cols
df.select_dtypes(include = 'object').columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordi = OrdinalEncoder(categories=[['Khordha','Cuttack', 'Puri', 'Bhubaneswar']])
df['City'] = ordi.fit_transform(df[['City']])
df['City']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Garage']=le.fit_transform(df['Garage'])
```

```python
from sklearn.preprocessing import OneHotEncoder
one=OneHotEncoder(sparse_output=False)
df1=one.fit_transform(df[['Property_Type']])
name=one.get_feature_names_out(['Property_Type'])
encod_df=pd.DataFrame(df1,columns=name)
df=pd.concat([df.drop('Property_Type',axis=1),encod_df],axis=1)
```

```python
df['Total_rooms']=df['Bedrooms']+df['Bedrooms']
```

```python
df['Price_Per_Sqft']=df['Price']/df['Area_sqft']
```

```python
df['House_Age']=2026-df['Year_Built']
```
